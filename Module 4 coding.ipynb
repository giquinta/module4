```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
page = requests.get('https://phoenix.craigslist.org/search/gms?sale_date=2022-02-06')
page.status_code
```

```python
bsobj = soup(page.content,'lxml')
bsobj
```

```python
links = []
for link in bsobj.findAll('li',{'class':'result-row'}):
    links.append(link.a['href'])
    
links
```

```python
title = []
moving = {'keyword':[],'url':[]}

for link in links:
    page = requests.get(link)
    bsobj = soup(page.content,'lxml')
    #print(bsobj.findAll('h1')[0].text.strip())
    title.append(bsobj.findAll('h1')[0].text.strip())
    text = ""
    for i in bsobj.findAll('section',{'id':'postingbody'}):
        text+=i.text.strip()
        #print(text.lower())
        if 'vacuums' in text.lower():
            moving['keyword'].append('vacuums')
            moving['url'].append(link)
        elif 'moving' in text.lower(): 
            moving['keyword'].append('moving')
            moving['url'].append(link)
        elif 'sale' in text.lower(): 
            moving['keyword'].append('sale')
            moving['url'].append(link)
        #if 'cabinet' in text: 
            #print(True)
```

```python
import pandas as pd
```

```python
dataframe = pd.DataFrame(moving)
```
